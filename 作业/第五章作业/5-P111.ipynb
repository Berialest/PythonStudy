```python
import pandas as pd
data=pd.read_csv('c:\\data\\wine.data',header=None)
data
```

```python
X=data.iloc[:,1:14].values.astype(float)
y=data.iloc[:,0].values.astype(int)
```

```python
from sklearn import model_selection
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier()
clf=clf.fit(X_train,y_train)
clf.score(X_train,y_train)
```

```python
predicted_y = clf.predict(X_test)
from sklearn import metrics
print(metrics.classification_report(y_test, predicted_y))
print('Confusion matrix:' )
print( metrics.confusion_matrix(y_test, predicted_y) )
```

```python
import pandas as pd
import numpy as np
from pandas import DataFrame,Series
```

```python
data=pd.read_excel('c:\\data\\housePrice.xlsx',index_col=0)
housep=data.iloc[0:51]
housep
```

```python
housep.loc[housep['建筑类型']=='板楼','建筑类型']=1
housep.loc[housep['建筑类型']=='塔楼','建筑类型']=0
housep.loc[housep['建筑结构']=='砖混结构','建筑结构']=0
housep.loc[housep['建筑结构']=='钢混结构','建筑结构']=1
housep.loc[housep['装修情况']=='简装','装修情况']=1
housep.loc[housep['装修情况']=='其他','装修情况']=2
housep.loc[housep['装修情况']=='精装','装修情况']=3
housep.loc[housep['装修情况']=='毛坯','装修情况']=0
housep.loc[housep['配备电梯']=='有','配备电梯']=1
housep.loc[housep['配备电梯']=='无','配备电梯']=0
housep.loc[housep['所处层数']=='高','所处层数']=1
housep.loc[housep['所处层数']=='中','所处层数']=2
housep.loc[housep['所处层数']=='低','所处层数']=3
housep
```

```python
import matplotlib.pyplot as plt
data.plot(kind='scatter',x='建筑面积(㎡)',y='总价(W)',title='总价和建筑面积的关系')
plt.xlabel("建筑面积(㎡)")
plt.ylabel("总价(W)")
plt.show()
```

```python
X=housep.iloc[:,[1,2,4,5,7,10,11,12,13,14,15,16,]].values.astype(float)
y=housep.iloc[:,1].values.astype(int)
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.3,random_state=1)
linregTr=LinearRegression()
linregTr.fit(X_train,y_train)
print(linregTr.coef_,linregTr.intercept_)
```

```python
from sklearn.externals import joblib
joblib.dump(linregTr,'c:\\data\\linreg.pkl')
load_linreg=joblib.load('c:\\data\\linreg.pkl')
new_x=np.array([[120,1,1,3,1,2,2,1,1,2010,1,20]])
print('预期价格：',load_linreg.predict(new_x))
```

```python
from sklearn import metrics
y_train_pred = linregTr.predict(X_train)
y_test_pred = linregTr.predict(X_test)
train_err = metrics.mean_squared_error(y_train, y_train_pred) 
test_err = metrics.mean_squared_error(y_test, y_test_pred) 
'The mean squar error of train and test are: {:.2f}, {:.2f}'.format(train_err, test_err) 
```

```python
predict_score =linregTr.score(X_test,y_test)
'The decision coeficient is:{:.2f} '.format(predict_score) 
```
