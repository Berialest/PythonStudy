```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('c:\\data\\bankpep.csv',index_col=0)
data
```

```python
data['age'].plot(kind='hist',bins=10,density=True,xlim=(0,100))
data['age'].plot(kind='kde')
plt.title('Customer Age')
plt.xlabel('Age')
plt.show()
```

```python
data.plot(kind='scatter',x='age',y='income',marker='s',label='(age,income)',grid=True)
plt.xlim(0,80)
plt.title('Customer Income')
plt.show()
```

```python
a=data[['age','income','children']]
pd.plotting.scatter_matrix(a)
plt.show()
```

```python
t4=data.groupby('region')
t4mean=t4['income'].mean()
t4mean
```

```python
t4std=t4['income'].std()
t4std
```

```python
t4mean.plot(kind='bar',yerr=t4std,rot=45)
plt.show()
```

```python
t51=data['sex'].value_counts()
t51
```

```python
fig=plt.figure(figsize=(6,6))
ax1=fig.add_subplot(2,2,1)
t51.plot(kind='pie',startangle=60,autopct='%.1f%%',ax=ax1)
plt.title('Customer Sex')

ax2=fig.add_subplot(2,2,2)
t52=data.loc[data['car']=='YES']
t52count=t52['sex'].value_counts()
t52count.plot(kind='pie',startangle=240,autopct='%.1f%%',ax=ax2)
plt.title('Customer Car Sex')

ax3=fig.add_subplot(2,2,3)
t53=data['children'].value_counts()
t53.plot(kind='pie',startangle=60,autopct='&.1f%%',ax=ax3)

plt.show()
```
