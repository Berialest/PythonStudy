```python
import matplotlib.pyplot as plt 
import pandas as pd
import pandas as pd
```

```python
stdata = pd.read_csv('c:\\data\\students.csv')     
stdata.plot(kind='scatter',x='Height',y='Weight',title='Students Body Shape', marker='*',grid=True, xlim=[150,200], ylim=[40,80], label='(Height,Weight)')   
plt.show()
```

```python
data1= stdata[stdata['Gender'] == 'male'] 
data2= stdata[stdata['Gender'] == 'female'] 
```

```python
plt.figure()
plt.scatter(data1['Height'],data1['Weight'],c='r',marker='s',label='Male')   
plt.scatter(data2['Height'],data2['Weight'],c='b',marker='^',label='Female') 
plt.xlim(150,200)                
plt.ylim(40,80)             
plt.title('Student Body')    
plt.xlabel('Weight')             
plt.ylabel('Height')           
plt.grid()                         
plt.legend(loc='upper right') 
plt.show()
```
