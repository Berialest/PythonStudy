```python
import matplotlib.pyplot as plt 
import pandas as pd
from pandas import DataFrame
from pandas import Series
```

```python
pcdi=[1.46,1.62,1.78,1.94,2.38,2.60]
data=DataFrame({'Income':pcdi},index=['2012','2013','2014','2015','2016','2017'])
data.plot()
plt.show()
```

```python
data.plot(title="Income chart",LineWidth=2,marker='s',linestyle='dashed',color='k',grid=True,alpha=0.9,use_index=True,yticks=[0.,0.5,1,1.,1.5,2.,2.5,3.])
plt.xticks(range(0,6),('2012','2013','2014','2015','2016','2017'))
plt.xlabel('Year',fontsize=12)
plt.ylabel('Income(RMB Ten Thousand)',fontsize=12)
plt.annotate('Largest!',xy=(5,2.60),xytext=(3,2.60),arrowprops=dict(arrowstyle='->'))
fig=plt.figure(figsize=(8,6))
ax1=fig.add_subplot(2,2,1)
data.plot(title="line chart",use_index=True,color='b',fontsize='small',ax=ax1)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Income',fontsize=12)
plt.xticks(range(0,6),('2012','','2014','','2016',''))
ax2=fig.add_subplot(2,1,2)   
data.plot(kind='bar',title='bar chart',use_index=True,fontsize='small',ax=ax2)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Income',fontsize=12)
ax3=fig.add_subplot(2,2,2)    
data.plot(kind='box',title='box-whisker plot',fontsize='small',xticks=[],ax=ax3) 
plt.xlabel('2012-2017',fontsize=12)
plt.ylabel('Income',fontsize=12)
```
