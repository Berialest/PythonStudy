```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('c:\\data\\high-speed rail.csv',index_col=0)
data
```

```python
data['Operation'].plot(kind='bar',title='Operation Mileage')
plt.ylabel('Mileage(km)')
plt.annotate('Longest!',xy=(0,20000),xytext=(1,20000),arrowprops=dict(arrowstyle='-'))
plt.show()
```

```python
data1=data[['Operation','Under-construction','Planning']]
data1.plot(kind='barh',stacked=True,title='Global trends of high-speed rail')
plt.xlabel('Mileage(km)')
plt.show()
```

```python
data['Operation'].plot(kind='pie',figsize=(6,6),title='Operation Mileage',fontsize=14,explode=[0.2,0,0,0,0,0],shadow=True,startangle=60, autopct='%1.1f%%')
plt.show()
```
